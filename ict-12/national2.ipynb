## 12-2. 国土数値情報 : 用途地域データの可視化

### 1. 北区の用途地域データの可視化
#### 1-a) 下準備

```python
import pandas as pd
import geopandas as gpd
from shapely.geometry import Polygon
import japanize_matplotlib
```

```python
KITA_YOUTO_PATH = 'data/nl/13117_youto.geojson'
```

#### 1-b) GeoPandasでの読み込み

```python
gdf_kita = gpd.read_file(KITA_YOUTO_PATH)
gdf_kita
```

#### 1-c) 用途地域での色分け

```python
gdf_kita.explore(column='用途地域')
```

```python
gdf_kita.plot(column='用途地域', figsize=(10, 10), legend=True)
```

### 2. 面積の確認
#### 2-a) 面積の算出

```python
gdf_kita['area'] = gdf_kita.to_crs(epsg=6691).area
gdf_kita
```

#### 2-b) 用途地域で集計

```python
gdf_kita.groupby('用途地域')[['area']].sum().sort_values('area', ascending=False)
```
