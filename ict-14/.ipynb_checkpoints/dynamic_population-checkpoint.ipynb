## 14-3. 人流データを触ってみよう

### 1. メッシュ属性データの読み込み
#### 1-a) 下準備

```python
import pandas as pd
import geopandas as gpd
from shapely.geometry import Polygon
```

```python
MESH_CSV = 'data/attribute_mesh1km_2020.csv'
```

#### 1-b) DataFrameへの読み込み

```python
df_mesh = pd.read_csv(MESH_CSV)
df_mesh
```

#### 1-c) 東京都への絞り込み

```python
df_mesh_tokyo = df_mesh[(df_mesh['prefcode'] == 13) & (df_mesh['lat_min'] > 35)]
df_mesh_tokyo
```

#### 1-d) ポリゴンの作成

```python
def create_polygon(row):
    return Polygon([
        (row['lon_min'], row['lat_min']),
        (row['lon_max'], row['lat_min']),
        (row['lon_max'], row['lat_max']),
        (row['lon_min'], row['lat_max'])
    ])

gdf_mesh_tokyo = gpd.GeoDataFrame(df_mesh_tokyo, geometry=df_mesh_tokyo.apply(create_polygon, axis=1), crs=4326)
gdf_mesh_tokyo
```

```python
gdf_mesh_tokyo = gdf_mesh_tokyo[['mesh1kmid', 'geometry']]
gdf_mesh_tokyo
```

```python
gdf_mesh_tokyo.plot()
```

### 2. 人流データの読み込み
#### 2-a) 2019年4月のデータの読み込み

```python
POPULATION_CSV = 'data/{}/{}/monthly_mdp_mesh1km.csv'
```

```python
df_population_19 = pd.read_csv(POPULATION_CSV.format('2019', '04'))
df_population_19
```

#### 2-b) メッシュデータとのマージ

```python
gdf_mesh_population_19 = gdf_mesh_tokyo.merge(df_population_19)
gdf_mesh_population_19
```

#### 2-c) 平日日中のデータを可視化

```python
gdf_mesh_population_19_day = gdf_mesh_population_19[(gdf_mesh_population_19['dayflag'] == 1) & (gdf_mesh_population_19['timezone'] == 0)]
gdf_mesh_population_19_day.explore(column='population', cmap='OrRd', vmax=50000)
```

#### 2-d) 平日夜間のデータを可視化

```python
gdf_mesh_population_19_night = gdf_mesh_population_19[(gdf_mesh_population_19['dayflag'] == 1) & (gdf_mesh_population_19['timezone'] == 1)]
gdf_mesh_population_19_night.explore(column='population', cmap='OrRd', vmax=50000)
```
