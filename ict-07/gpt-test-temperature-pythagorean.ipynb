```python
#Cell_1.
import openai
import re
import os
from openai import OpenAI

oai = OpenAI(
    api_key = 'xxxx',
    base_url = "https://api.openai.iniad.org/api/v1"
)
```

```python
#Cell_2.
temperatures = [0, 0, 0]
sysmsg ="あなたは数学者です"
text = """ピタゴラスの定理の証明をひとつ教えてください。
 """

messages=[
    {'role': 'system', 'content': sysmsg},
    {'role': 'user', 'content': text},
]
i=0
for temp in temperatures:
    i+=1
    print("第" + str(i) + "回目の回答です。")
    resp = oai.chat.completions.create(
        model='gpt-4o-mini',
        messages=messages,
        temperature=temp,
    )
    
    print(resp.choices[0].message.content)
```

```python
#Cell_3.
temperatures = [1.7, 1.9, 1.9]
# Main loop
while True:
    # Get user input
    sysmsg ="あなたは数学者です"
    question = input("Enter your question (or 'q' to quit): ")

    # Check if the user wants to quit
    if question.lower() == 'q':
        break
        
    text = question

    messages=[
        {'role': 'system', 'content': sysmsg},
        {'role': 'user', 'content': text},
    ]

    i=0
    for temp in temperatures:
        i+=1
        print("第" + str(i) + "回目の回答です。")
        resp = oai.chat.completions.create(
            model='gpt-4o-mini',
            messages=messages,
            temperature=temp,
        )
        
        print(resp.choices[0].message.content)
```
