```python
#Cell_1.
import openai
import re
import os
from openai import OpenAI

oai = OpenAI(
    api_key = '',
    base_url = "https://api.openai.iniad.org/api/v1"
)
```

```python
#Cell_2.

messages=[
       {"role": "system", "content": "あなたはデータサイエンティストです。"},
       {"role": "user", "content": "以下のURLをスクレイピングするためのPythonコードを教えてください。https://quotes.toscrape.com/"}
]

resp = oai.chat.completions.create(
    model='gpt-4o-mini',
    messages=messages,
)

print(resp.choices[0].message.content)
```

```python
#Cell_3.
import requests
from bs4 import BeautifulSoup

# スクレイピングするURL
url = "https://quotes.toscrape.com/"

# requestsを使ってURLの内容を取得
response = requests.get(url)

# 確認のため、レスポンスのステータスコードを表示
if response.status_code == 200:
    # BeautifulSoupを使ってHTMLの解析
    soup = BeautifulSoup(response.text, 'html.parser')
    
    # 名言を格納するリスト
    quotes = []

    # 名言と著者を取得
    for quote in soup.find_all('div', class_='quote'):
        text = quote.find('span', class_='text').get_text()
        author = quote.find('small', class_='author').get_text()
        quotes.append({'text': text, 'author': author})

    # 名言を表示
    for q in quotes:
        print(f"{q['text']} — {q['author']}")
else:
    print(f"Error: {response.status_code}")
```

```python
#Cell_5. #Slightly changed.
import requests
from bs4 import BeautifulSoup

# URLを指定
url = 'https://quotes.toscrape.com/'

# URLにリクエストを送信
response = requests.get(url)
quotes_all=[]

# ステータスコードを確認
if response.status_code == 200:
    # コンテンツをBeautifulSoupで解析
    soup = BeautifulSoup(response.text, 'html.parser')

    # 名言を抽出
    quotes = soup.find_all('div', class_='quote')
    #print(quotes)
    for quote in quotes:
        text = quote.find('span', class_='text').get_text()
        author = quote.find('small', class_='author').get_text()
        print(f'Quote: {text}\nAuthor: {author}\n')
        quotes_all.append({'text': text, 'author': author})    

else:
    print(f'Failed to retrieve page with status code: {response.status_code}')
```
