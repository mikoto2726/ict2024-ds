```python
#Cell_1.
import openai
import re
import os
from openai import OpenAI

oai = OpenAI(
    api_key = '',
    base_url = "https://api.openai.iniad.org/api/v1"
)
```

```python
#Cell_2.
messages=[
       {"role": "system", "content": "あなたはデータサイエンティストです。"},
       {"role": "user", "content": "以下のURLをスクレイピングするためのPythonコードを教えてください。https://books.toscrape.com/"}
]

resp = oai.chat.completions.create(
    model='gpt-4o-mini',
    messages=messages,
)

print(resp.choices[0].message.content)
```

```python
#Cell_3. #As ChatGPT said:
import requests
from bs4 import BeautifulSoup

# スクレイピングするURL
url = 'https://books.toscrape.com/'

# URLにリクエストを送る
response = requests.get(url)

# レスポンスが成功したか確認
if response.status_code == 200:
    # BeautifulSoupでHTMLを解析
    soup = BeautifulSoup(response.text, 'html.parser')
    
    # 書籍情報を格納するリスト
    books = []

    # 各書籍ブロックを取得
    for book in soup.find_all('article', class_='product_pod'):
        title = book.h3.a['title']  # 書籍タイトル
        price = book.find('p', class_='price_color').text  # 価格
        books.append({'title': title, 'price': price})

    # 結果を表示
    for book in books:
        print(f"Title: {book['title']}, Price: {book['price']}")
else:
    print(f"Failed to retrieve data: {response.status_code}")
```

```python
#Cell_5. #Slightly changed.
import requests
from bs4 import BeautifulSoup

# ターゲットURL
url = 'https://books.toscrape.com/'

# ページの内容を取得
response = requests.get(url)
books = []
# レスポンスが成功したか確認
if response.status_code == 200:
    # BeautifulSoupで解析
    soup = BeautifulSoup(response.text, 'html.parser')


    # 本のタイトルと価格を抽出
    for book in soup.find_all('article', class_='product_pod'):
        title = book.find('h3').find('a')['title']
        price = book.find('p', class_='price_color').text
        books.append({'title': title, 'price': price})

    # 本の情報を表示
    for book in books:
        print(f"Title: {book['title']}, Price: {book['price']}")
else:
    print(f"Failed to retrieve the webpage. Status code: {response.status_code}")
```
